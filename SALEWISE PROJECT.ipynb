```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, Markdown

print("Starting SalesWise AI Agent...")


print("\n Loading data...")
try:
    df = pd.read_csv(r"D:\Semester 5\Data visooo\project1\train.csv")
    
    # Convert dates
    df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')
    df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d/%m/%Y')
    
    # Extract date features
    df['Order Year'] = df['Order Date'].dt.year
    df['Order Month'] = df['Order Date'].dt.month
    df['Order Quarter'] = df['Order Date'].dt.quarter
    df['Order Day'] = df['Order Date'].dt.day
    df['Order Weekday'] = df['Order Date'].dt.day_name()
    
    # Calculate shipping time
    df['Shipping Time'] = (df['Ship Date'] - df['Order Date']).dt.days
    
    print(f" Successfully loaded {len(df)} records")
    print(f" Data range: {df['Order Date'].min().date()} to {df['Order Date'].max().date()}")
    
except Exception as e:
    print(f"Error loading data: {e}")
    raise

print(" EXECUTIVE DASHBOARD - KEY INSIGHTS")

total_sales = df['Sales'].sum()
total_orders = df['Order ID'].nunique()
avg_order = total_sales / total_orders
best_category = df.groupby('Category')['Sales'].sum().idxmax()
best_region = df.groupby('Region')['Sales'].sum().idxmax()

print(f"""
 TOTAL SALES:        ${total_sales:,.2f}
 TOTAL ORDERS:       {total_orders:,}
 AVERAGE ORDER:      ${avg_order:,.2f}
 BEST CATEGORY:      {best_category}
 BEST REGION:        {best_region}
""")

print("📊 BASIC STATISTICS")

stats = {
    'Total Sales': f"${df['Sales'].sum():,.2f}",
    'Total Orders': df['Order ID'].nunique(),
    'Total Customers': df['Customer ID'].nunique(),
    'Total Products': df['Product ID'].nunique(),
    'Date Range': f"{df['Order Date'].min().date()} to {df['Order Date'].max().date()}",
    'Average Order Value': f"${df['Sales'].mean():,.2f}",
    'Max Order Value': f"${df['Sales'].max():,.2f}",
    'Min Order Value': f"${df['Sales'].min():,.2f}"
}

for key, value in stats.items():
    print(f"• {key}: {value}")

print(" CATEGORY PERFORMANCE")

category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
category_orders = df.groupby('Category')['Order ID'].nunique()

for i, (category, sales) in enumerate(category_sales.items(), 1):
    orders = category_orders[category]
    avg_sale = sales / orders if orders > 0 else 0
    print(f"{i}. {category}:")
    print(f"   Sales: ${sales:,.2f}")
    print(f"   Orders: {orders}")
    print(f"   Average per order: ${avg_sale:,.2f}")
    print()

# Visualization
plt.figure(figsize=(10, 5))
category_sales.plot(kind='bar', color=['#2E86AB', '#A23B72', '#F18F01'])
plt.title('Total Sales by Category', fontsize=14)
plt.ylabel('Sales ($)', fontsize=12)
plt.xlabel('Category', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

print(" REGIONAL PERFORMANCE")

region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)

for i, (region, sales) in enumerate(region_sales.items(), 1):
    percentage = (sales / df['Sales'].sum()) * 100
    print(f"{i}. {region}: ${sales:,.2f} ({percentage:.1f}%)")

# Visualization
plt.figure(figsize=(8, 6))
region_sales.plot(kind='pie', autopct='%1.1f%%', colors=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4'])
plt.title('Sales Distribution by Region', fontsize=14)
plt.ylabel('')
plt.tight_layout()
plt.show()

print(" HEATMAP: Sales by Category and Region")

# Create pivot table
heatmap_data = df.pivot_table(values='Sales', index='Category', columns='Region', aggfunc='sum')

# Display as formatted table
print("\nSales Matrix ($):")
for cat in heatmap_data.index:
    row = [f"{cat:15}"] + [f"${heatmap_data.loc[cat, reg]:,.0f}" for reg in heatmap_data.columns]
    print(" | ".join(row))

# Visualization
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, fmt='.0f', cmap='YlOrRd', linewidths=0.5)
plt.title('Sales Heatmap: Category vs Region', fontsize=14)
plt.tight_layout()
plt.show()

print("TOP 10 PRODUCTS")

top_products = df.groupby('Product Name').agg({
    'Sales': 'sum',
    'Order ID': 'count'
}).nlargest(10, 'Sales').sort_values('Sales', ascending=False)

top_products['Avg Price'] = top_products['Sales'] / top_products['Order ID']

for i, (product, row) in enumerate(top_products.iterrows(), 1):
    print(f"{i}. {product[:50]}...")
    print(f"   Total Sales: ${row['Sales']:,.2f}")
    print(f"   Orders: {row['Order ID']}")
    print(f"   Average Price: ${row['Avg Price']:,.2f}")
    print()

print(" TOP 10 CUSTOMERS")

top_customers = df.groupby('Customer Name').agg({
    'Sales': 'sum',
    'Order ID': 'count',
    'Segment': 'first'
}).nlargest(10, 'Sales').sort_values('Sales', ascending=False)

for i, (customer, row) in enumerate(top_customers.iterrows(), 1):
    print(f"{i}. {customer}")
    print(f"   Total Spent: ${row['Sales']:,.2f}")
    print(f"   Orders: {row['Order ID']}")
    print(f"   Segment: {row['Segment']}")
    print()

print(" MONTHLY SALES TREND")

# Create year-month column
df['Year-Month'] = df['Order Date'].dt.to_period('M')
monthly_sales = df.groupby('Year-Month')['Sales'].sum()

# Convert to readable format
monthly_sales_df = monthly_sales.reset_index()
monthly_sales_df['Year-Month'] = monthly_sales_df['Year-Month'].astype(str)

# Calculate growth
monthly_sales_df['Growth'] = monthly_sales_df['Sales'].pct_change() * 100

print("Monthly Sales Report:")
for _, row in monthly_sales_df.iterrows():
    growth_symbol = "📈" if row['Growth'] > 0 else "📉" if pd.notna(row['Growth']) else ""
    growth_text = f"{growth_symbol} {row['Growth']:+.1f}%" if pd.notna(row['Growth']) else "N/A"
    print(f"{row['Year-Month']}: ${row['Sales']:,.2f} ({growth_text})")

# Visualization
plt.figure(figsize=(12, 5))
plt.plot(monthly_sales_df['Year-Month'], monthly_sales_df['Sales'], marker='o', linewidth=2, color='#2E86AB')
plt.title('Monthly Sales Trend', fontsize=14)
plt.ylabel('Sales ($)', fontsize=12)
plt.xlabel('Month', fontsize=12)
plt.xticks(rotation=45)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

print(" SEASONALITY ANALYSIS - Best & Worst Months")

monthly_avg = df.groupby('Order Month')['Sales'].mean()
best_month = monthly_avg.idxmax()
worst_month = monthly_avg.idxmin()
best_month_sales = monthly_avg.max()
worst_month_sales = monthly_avg.min()

# Create month names
month_names = {1:'Jan', 2:'Feb', 3:'Mar', 4:'Apr', 5:'May', 6:'Jun',
               7:'Jul', 8:'Aug', 9:'Sep', 10:'Oct', 11:'Nov', 12:'Dec'}

print(f"""
 BEST MONTH:         {month_names[best_month]} (Month {best_month}) - ${best_month_sales:,.2f} avg per order
 WORST MONTH:        {month_names[worst_month]} (Month {worst_month}) - ${worst_month_sales:,.2f} avg per order

 Seasonal Pattern:
   • Q4 (Oct-Dec) shows highest sales (holiday season)
   • Q1 (Jan-Mar) typically lowest
""")

# Visualization
plt.figure(figsize=(10, 5))
monthly_avg.plot(kind='bar', color='#F18F01')
plt.title('Average Sales by Month', fontsize=14)
plt.ylabel('Average Sales ($)', fontsize=12)
plt.xlabel('Month', fontsize=12)
plt.xticks(ticks=range(12), labels=[month_names[i] for i in range(1,13)], rotation=0)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

print(" SIMPLE FORECAST - Next Month Sales Prediction")

# Calculate last 3 months average growth
last_3_months = monthly_sales.tail(3)
avg_growth = last_3_months.pct_change().mean()
last_month_sales = monthly_sales.iloc[-1]
forecast = last_month_sales * (1 + avg_growth)

print(f"""
 Last month sales:     ${last_month_sales:,.2f}
 Avg growth (3 months): {avg_growth*100:.1f}%
 Forecast next month:   ${forecast:,.2f}

 Note: Simple forecast based on recent trend only.
""")

print("SHIPPING ANALYSIS")

# Shipping mode analysis
shipping_analysis = df.groupby('Ship Mode').agg({
    'Sales': 'sum',
    'Order ID': 'count',
    'Shipping Time': 'mean'
})

print("Shipping Mode Performance:")
print("-" * 40)
for mode, row in shipping_analysis.iterrows():
    avg_shipping_time = row['Shipping Time']
    avg_order_value = row['Sales'] / row['Order ID']
    print(f"• {mode}:")
    print(f"  Orders: {row['Order ID']}")
    print(f"  Total Sales: ${row['Sales']:,.2f}")
    print(f"  Avg Order Value: ${avg_order_value:,.2f}")
    print(f"  Avg Shipping Time: {avg_shipping_time:.1f} days")
    print()

print(" CUSTOMER SEGMENT ANALYSIS")

segment_analysis = df.groupby('Segment').agg({
    'Sales': 'sum',
    'Customer ID': 'nunique',
    'Order ID': 'nunique'
})

segment_analysis['Sales per Customer'] = segment_analysis['Sales'] / segment_analysis['Customer ID']
segment_analysis['Orders per Customer'] = segment_analysis['Order ID'] / segment_analysis['Customer ID']

for segment, row in segment_analysis.iterrows():
    print(f"• {segment} Segment:")
    print(f"  Total Customers: {row['Customer ID']}")
    print(f"  Total Sales: ${row['Sales']:,.2f}")
    print(f"  Sales per Customer: ${row['Sales per Customer']:,.2f}")
    print(f"  Orders per Customer: {row['Orders per Customer']:.1f}")
    print()

print("KEY BUSINESS INSIGHTS")

insights = []

# Insight 1: Best performing category
best_category = category_sales.idxmax()
best_category_sales = category_sales.max()
category_percentage = (best_category_sales / df['Sales'].sum()) * 100
insights.append(f"1. **{best_category}** is the best performing category with ${best_category_sales:,.2f} ({category_percentage:.1f}% of total sales)")

# Insight 2: Top region
best_region = region_sales.idxmax()
best_region_sales = region_sales.max()
region_percentage = (best_region_sales / df['Sales'].sum()) * 100
insights.append(f"2. **{best_region}** is the top region with ${best_region_sales:,.2f} ({region_percentage:.1f}% of total sales)")

# Insight 3: Customer concentration
top_5_customers = df.groupby('Customer Name')['Sales'].sum().nlargest(5).sum()
customer_concentration = (top_5_customers / df['Sales'].sum()) * 100
insights.append(f"3. **Customer Concentration**: Top 5 customers contribute {customer_concentration:.1f}% of total sales")

# Insight 4: Shipping time
avg_shipping_time = df['Shipping Time'].mean()
insights.append(f"4. **Shipping Performance**: Average shipping time is {avg_shipping_time:.1f} days")

# Insight 5: Growth trend
if len(monthly_sales_df) > 1:
    latest_growth = monthly_sales_df['Growth'].iloc[-1]
    growth_trend = "growing" if latest_growth > 0 else "declining"
    insights.append(f"5. **Sales Trend**: Recent month shows {latest_growth:+.1f}% growth (business is {growth_trend})")

# Insight 6: Seasonality
insights.append(f"6. **Seasonality**: Best month is {month_names[best_month]} (${best_month_sales:,.2f}), worst month is {month_names[worst_month]} (${worst_month_sales:,.2f})")

# Insight 7: Forecast
insights.append(f"7. **Forecast**: Next month projected sales: ${forecast:,.2f} (based on {avg_growth*100:.1f}% avg growth)")

# Print all insights
for insight in insights:
    print(f"• {insight}")

print(" AI AGENT - ASK QUESTIONS")

def ask_sales_agent(question, data_frame):
    """AI agent for sales data analysis with enhanced capabilities"""
    
    question_lower = question.lower()
    
    if 'total sales' in question_lower:
        total = data_frame['Sales'].sum()
        return f" Total sales: ${total:,.2f}"
    
    elif 'best category' in question_lower or 'top category' in question_lower:
        category_sales = data_frame.groupby('Category')['Sales'].sum()
        best = category_sales.idxmax()
        amount = category_sales.max()
        return f" Best category: **{best}** with ${amount:,.2f} in sales"
    
    elif 'best region' in question_lower:
        region_sales = data_frame.groupby('Region')['Sales'].sum()
        best = region_sales.idxmax()
        amount = region_sales.max()
        return f" Best region: **{best}** with ${amount:,.2f} in sales"
    
    elif 'top product' in question_lower:
        top_product = data_frame.groupby('Product Name')['Sales'].sum().nlargest(1)
        product = top_product.index[0]
        amount = top_product.iloc[0]
        return f" Top product: **{product[:50]}...** with ${amount:,.2f}"
    
    elif 'average order' in question_lower:
        avg = data_frame['Sales'].mean()
        return f" Average order value: ${avg:,.2f}"
    
    elif 'growth' in question_lower:
        if len(monthly_sales_df) > 1:
            latest_growth = monthly_sales_df['Growth'].iloc[-1]
            return f" Latest monthly growth rate: {latest_growth:+.1f}%"
        else:
            return " Not enough data for growth analysis"
    
    elif 'forecast' in question_lower or 'predict' in question_lower:
        last_3_months = monthly_sales.tail(3)
        avg_growth = last_3_months.pct_change().mean()
        last_month_sales_val = monthly_sales.iloc[-1]
        forecast_val = last_month_sales_val * (1 + avg_growth)
        return f" Next month sales forecast: ${forecast_val:,.2f} (based on {avg_growth*100:.1f}% avg growth)"
    
    elif 'season' in question_lower or 'best month' in question_lower:
        monthly_avg = data_frame.groupby('Order Month')['Sales'].mean()
        best = monthly_avg.idxmax()
        best_val = monthly_avg[best]
        return f" Best month is **{month_names[best]} (Month {best})** with ${best_val:,.2f} avg per order – typically holiday season!"
    
    elif 'worst month' in question_lower:
        monthly_avg = data_frame.groupby('Order Month')['Sales'].mean()
        worst = monthly_avg.idxmin()
        worst_val = monthly_avg[worst]
        return f" Worst month is **{month_names[worst]} (Month {worst})** with ${worst_val:,.2f} avg per order"
    
    elif 'heatmap' in question_lower:
        return " Check the heatmap above! It shows Technology dominates West ($308K) and East ($254K) regions."
    
    elif 'customer' in question_lower and 'total' in question_lower:
        total_customers = data_frame['Customer ID'].nunique()
        return f"Total unique customers: {total_customers}"
    
    elif 'help' in question_lower:
        return """
         **I can answer questions about:**
        • Total sales
        • Best category
        • Best region
        • Top products
        • Average order value
        • Sales growth
        • Customer count
        •  **NEW** - Sales forecast (next month)
        •  **NEW** - Best/worst month (seasonality)
        •  **NEW** - Heatmap analysis
        • And more!
        
        Try asking:
        - "What is the forecast for next month?"
        - "Which is the best month for sales?"
        - "Show me the worst month"
        - "What does the heatmap show?"
        """
    
    else:
        return " I'm not sure about that. Try asking about total sales, best category, forecast, or seasonality. Type 'help' for options."

# Interactive Q&A
print("Try asking me questions about the sales data!")
print("Examples:")
print("- 'What are the total sales?'")
print("- 'Which is the best category?'")
print("- 'What is the forecast for next month?'")
print("- 'Which is the best month for sales?'")
print("- 'What does the heatmap show?'")
print("- 'help' for all options")
print()

# Interactive Q&A loop
while True:
    user_question = input(" Your question (or 'quit' to exit): ").strip()
    
    if user_question.lower() in ['quit', 'exit', 'q']:
        print("\n Thank you for using SalesWise AI Agent!")
        break
    
    if user_question:
        answer = ask_sales_agent(user_question, df)
        print(f"\n AI Agent: {answer}")

# ==================== END ====================
print(" SalesWise AI Agent analysis completed!")
print("\n Generated analysis includes:")
print("• Executive Dashboard with KPIs")
print("• Category & Regional performance")
print("• Heatmap (Category vs Region)")
print("• Top products & customers")
print("• Monthly sales trends")
print("• Seasonality analysis (best/worst months)")
print("• Sales forecast for next month")
print("• Shipping analysis")
print("• Customer segments")
print("• Business insights")
print("• Interactive Q&A agent with enhanced capabilities")
print("\n All done! Ready for your next analysis.")
```
